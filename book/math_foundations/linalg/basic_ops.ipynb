# Element-gewijze operaties
Hier beschrijven we de basis operaties optellen/aftrekken en vermenigvuldigen/delen (producten) waarbij de operatie _element-gewijs_ wordt toegepast.
We zullen later zien dat producten tussen tensors speciale vormen kunnen aannemen.

:::{important}
Bij element-gewijze operaties tussen tensors impliceert dit dat de dimensionaliteit moet kloppen. Denk aan lego blokken die op elkaar moeten passen.
:::

## Scalair optellen
$$
\pmb{a} + b = \begin{bmatrix}
a_1 + b \cr
a_2 +b \cr
\vdots  \cr
a_n + b
\end{bmatrix}
$$

$$
\pmb{A} + b = \begin{bmatrix}
A_{1,1} + b & \ldots & A_{1,j} + b & \ldots & A_{1,n} + b \cr
\vdots  &  & \vdots &  & \vdots \cr
A_{i,1} + b & \ldots & A_{i,j} + b & \ldots & A_{i,n} + b \cr
\vdots  &  & \vdots &  & \vdots \cr
A_{m,1} + b & \ldots & A_{m,j} + b & \ldots & A_{m,n} + b
\end{bmatrix}
$$

## Scalaire vermenigvuldiging
$$
\pmb{a}b = \begin{bmatrix}
a_1b \cr
a_2b \cr
\vdots  \cr
a_nb
\end{bmatrix}
$$

$$
\pmb{A}b = \begin{bmatrix}
A_{1,1}b & \ldots & A_{1,j}b & \ldots & A_{1,n}b \cr
\vdots  &  & \vdots &  & \vdots \cr
A_{i,1}b & \ldots & A_{i,j}b & \ldots & A_{i,n}b \cr
\vdots  &  & \vdots &  & \vdots \cr
A_{m,1}b & \ldots & A_{m,j}b & \ldots & A_{m,n}b
\end{bmatrix}
$$

## Tensors optellen
:::{warning} $ $
Enkel bij passende dimensies.
:::
als
$$
\pmb{a} = \begin{bmatrix}
a_1 \cr
a_2 \cr
\vdots  \cr
a_n
\end{bmatrix}, \pmb{b} = \begin{bmatrix}
b_1 \cr
b_2 \cr
\vdots  \cr
b_n
\end{bmatrix}
$$
dan
$$
\pmb{a} + \pmb{b} = \begin{bmatrix}
a_1 + b_1 \cr
a_2 + b_2 \cr
\vdots  \cr
a_n + b_n
\end{bmatrix}
$$

als
$$
\pmb{A} = \begin{bmatrix}
A_{1,1} & \ldots & A_{1,j} & \ldots & A_{1,n} \cr
\vdots  &  & \vdots &  & \vdots \cr
A_{i,1} & \ldots & A_{i,j} & \ldots & A_{i,n} \cr
\vdots  &  & \vdots &  & \vdots \cr
A_{m,1} & \ldots & A_{m,j} & \ldots & A_{m,n}
\end{bmatrix}, \pmb{B} = \begin{bmatrix}
B_{1,1} & \ldots & B_{1,j} & \ldots & B_{1,n} \cr
\vdots  &  & \vdots &  & \vdots \cr
B_{i,1} & \ldots & B_{i,j} & \ldots & B_{i,n} \cr
\vdots  &  & \vdots &  & \vdots \cr
B_{m,1} & \ldots & B_{m,j} & \ldots & B_{m,n}
\end{bmatrix}
$$
dan
$$
\pmb{A} + \pmb{B} = \begin{bmatrix}
A_{1,1} + B_{1,1} & \ldots & A_{1,j} + B_{1,j} & \ldots & A_{1,n} + B_{1,n}\cr
\vdots  &  & \vdots &  & \vdots \cr
A_{i,1} + B_{i,1} & \ldots & A_{i,j} + B_{i,j} & \ldots & A_{i,n} + B_{i,n} \cr
\vdots  &  & \vdots &  & \vdots \cr
A_{m,1} + B_{m,1} & \ldots & A_{m,j} + B_{m,j} & \ldots & A_{m,n} + B_{m,n}
\end{bmatrix}
$$
en
$$
\pmb{A}^T + \pmb{b} = \begin{bmatrix}
A_{1,1} + b_1 & \ldots & A_{1, i} + b_1 & \ldots & A_{1, m} + b_1 \cr
\vdots  &  & \vdots &  & \vdots \cr
A_{j, 1} + b_2 & \ldots & A_{j,i} + b_2 & \ldots & A_{j,m} + b_2 \cr
\vdots  &  & \vdots &  & \vdots \cr
A_{n,1} + b_n & \ldots & A_{n,i} + b_n & \ldots & A_{n,m} + b_n
\end{bmatrix}
$$

## Element-gewijs product
Dit is het zogenaamde _Hadamard_ product. Let op het speciale symbool voor de operatie!

$$
\pmb{a} \odot \pmb{b} = \begin{bmatrix}
a_1 \times b_1 \cr
a_2 \times b_2 \cr
\vdots  \cr
a_n \times b_n
\end{bmatrix}
$$

$$
\pmb{A} \odot \pmb{B} = \begin{bmatrix}
A_{1,1} \times B_{1,1} & \ldots & A_{1,j} \times B_{1,j} & \ldots & A_{1,n} \times B_{1,n}\cr
\vdots  &  & \vdots &  & \vdots \cr
A_{i,1} \times B_{i,1} & \ldots & A_{i,j} \times B_{i,j} & \ldots & A_{i,n} \times B_{i,n} \cr
\vdots  &  & \vdots &  & \vdots \cr
A_{m,1} \times B_{m,1} & \ldots & A_{m,j} \times B_{m,j} & \ldots & A_{m,n} \times B_{m,n}
\end{bmatrix}
$$

$$
\pmb{A}^T \odot \pmb{b} = \begin{bmatrix}
A_{1,1}b_1 & \ldots & A_{1, i}b_1 & \ldots & A_{1, m}b_1 \cr
\vdots  &  & \vdots &  & \vdots \cr
A_{j, 1}b_2 & \ldots & A_{j,i}b_2 & \ldots & A_{j,m}b_2 \cr
\vdots  &  & \vdots &  & \vdots \cr
A_{n,1}b_n & \ldots & A_{n,i}b_n & \ldots & A_{n,m}b_n
\end{bmatrix}
$$
