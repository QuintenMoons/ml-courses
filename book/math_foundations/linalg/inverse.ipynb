# Matrix inverse

Matrix inversie is een sleutel tot belangrijke mogelijkheden in Machine Learning. In de vorige sectie zagen we de **identiteitsmatrix** als speciale matrix.
Deze spelen een centrale rol in de formele definitie van de **matrix inverse** $\pmb{A}^{-1}$ van een matrix $\pmb{A}$:

$$
\pmb{A}^{-1}\pmb{A} = \pmb{I}_n
$$

(target-matrix-inverse)=
**De matrix inverse $\pmb{A}^{-1}$ is de matrix die bij vermenigvuldiging met \pmb{A} de identiteitsmatrix \pmb{I}_n oplevert**.

## Stelsels van lineaire vergelijkingen
We kunnen matrix inverses gebruiken om oplossingen te vinden voor zogenaamde _stelsels van lineaire vergelijkingen_. De algemene vorm van een lineaire vergelijking is:
$$
y = \beta_1x_1 + \beta_2x_2 + \ldots + \beta_nx_n
$$
We kunnen dit in vector notatie vertalen naar:
$$
y = \begin{bmatrix} \beta_1 & \beta_2 & \ldots & \beta_n\end{bmatrix}^T\begin{bmatrix}
x_1 \cr
x_2 \cr
\vdots \cr
x_n
\end{bmatrix}
$$
Bij een stelsel van lineaire vergelijkingen blijft $\pmb{b} = \begin{bmatrix} \beta_1 & \beta_2 & \ldots & \beta_n\end{bmatrix}^T$ constant, maar zijn er verschillende instanties van
$\pmb{x} = \begin{bmatrix} x_1 & x_2 & \ldots & x_n\end{bmatrix}^T$:

$$
\begin{align}
y_1 &= \beta_1x_{1,1} + \beta_2x_{1,2} + \ldots + \beta_nx_{1,n} \cr
y_2 &= \beta_1x_{2,1} + \beta_2x_{2,2} + \ldots + \beta_nx_{2,n} \cr
\ldots \cr
y_m &= \beta_1x_{m,1} + \beta_2x_{m,2} + \ldots + \beta_nx_{m,n} \cr
\end{align}
$$
Dit kunnen compact in matrix notatie schrijven:
$$
\pmb{y} = \pmb{X}\pmb{b}
$$
In een stelsel van vergelijkingen is het de bedoeling om een oplossing te vinden voor $\pmb{b}$:

$$
\begin{align}
\pmb{y} &= \pmb{X}\pmb{b} \cr
\pmb{X}^{-1}\pmb{y} &= \pmb{X}^{-1}\pmb{X}\pmb{b} \cr
\pmb{X}^{-1}\pmb{y} &= \pmb{I}_n\pmb{b} \cr
\pmb{X}^{-1}\pmb{y} &= \pmb{b} \cr
\pmb{b} &= \pmb{X}^{-1}\pmb{y}
\end{align}
$$

We kunnen de vergelijkingen dus oplossen voor $\pmb{b}$ als we de inverse $\pmb{X}^{-1}$ kunnen berekenen.
Dat is helaas minder evident dan we zouden hopen, maar dat is voor een volgende sectie.
Hier geven we een eenvoudig voorbeeld met een stelsel van 2 vergelijkingen en $n=2$.

$$
\begin{align}
-5 &= \beta_1 + 3\beta_2 \cr
4 &= 2\beta_1 - \beta_2
\end{align}
$$

We hebben dus:

$$
\begin{align}
\pmb{y} &= \begin{bmatrix}
-5 & 4
\end{bmatrix}^T \cr
\pmb{X} &= \begin{bmatrix}
1 & 3 \cr
2 & -1
\end{bmatrix}
\end{align}
$$

Om $\pmb{b}$ te vinden zoeken we de inverse van $\pmb{X}$ die voor het $2 \times 2$ geval de volgende speciale vorm aanneemt:

$$
\begin{align}
\pmb{X}^{-1} &= \frac{1}{det(\pmb{X})}\begin{bmatrix}
x_{2,2} & -x_{1,2} \cr
-x_{2,1} & x_{1,1}
\end{bmatrix} \cr
&= \frac{1}{
    x_{1,1}x_{2,2} - x_{1,2}x_{2,1}
}\begin{bmatrix}
x_{2,2} & -x_{1,2} \cr
-x_{2,1} & x_{1,1}
\end{bmatrix}
\end{align}
$$

We krijgen bijgevolg:

$$
\begin{align}
\pmb{X}^{-1} &= \frac{1}{
    -1 - 3 \times 2
}\begin{bmatrix}
-1 & -3 \cr
-2 & 1
\end{bmatrix} \cr
&= \frac{1}{-7}\begin{bmatrix}
-1 & -3 \cr
-2 & 1
\end{bmatrix} \cr
&= \begin{bmatrix}
\frac{1}{7} & \frac{3}{7} \cr
\frac{2}{7} & -\frac{1}{7}
\end{bmatrix}
\end{align}
$$

en voor de uiteindelijke oplossing $\pmb{b} = \pmb{X}^{-1}\pmb{y}$

$$
\begin{align}
\pmb{b} &= \begin{bmatrix}
\frac{1}{7} & \frac{3}{7} \cr
\frac{2}{7} & -\frac{1}{7}
\end{bmatrix} \begin{bmatrix}
-5 \cr
4
\end{bmatrix} \cr
&= \begin{bmatrix}
\frac{1}{7}(-5) + \frac{3}{7}(4) \cr
\frac{2}{7}(-5) + (-\frac{1}{7})(4)
\end{bmatrix} \cr
&= \begin{bmatrix}
1 \cr
-2
\end{bmatrix}
\end{align}
$$

Ter verificatie:

$$
\begin{align}
-5 &= (1) + (-2) \times 3 = 1 - 6 = -5 \cr
4 &= (1) \times 2 - (-2) = 2 + 2 = 4
\end{align}
$$
